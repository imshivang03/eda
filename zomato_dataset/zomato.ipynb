```python
#combining the both dataframes using country_code feature
final_df= pd.merge(df, df_country, on= "Country Code", how= "left")
```

```python
# Zomato is mostly available in India, Zomato is basically earning more in India
final_df.Country.value_counts()
```

```python
country_names= final_df.Country.value_counts().index
```

```python
country_val= final_df.Country.value_counts().values
```

```python
# Pie chart- top 3 countries that uses zomato
plt.pie(country_val[:3], labels= country_names[:3], autopct= "%1.2f%%")
```

Observation: India has the maximum number of orders, India is the business hub for Zomato and then USA is on the 2nd position in terms of placing the orders on zomato and UK is on the 3rd position. 

```python
final_df.columns
```

```python
ratings= final_df.groupby(["Aggregate rating", "Rating color", "Rating text"]).size().reset_index().rename(columns= {0:"Rating count"})
```

```python
ratings
```

# Observations
1. Maximum number of people didn't give ratings
2. When Rating is between 4.5 to 4.9 ---> Excellent
3. When Rating is between 4.0 to 4.4 ---> Very Good
4. When rating is between 3.5 to 3.9 ---> Good
5. When rating is between 2.5 to 3.4 ---> Average
6. When rating is between 1.8 to 2.4 ---> Poor
7. When rating is 0.0 ---> Not Rated

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams["figure.figsize"]= (12, 6)
sns.barplot(x= "Aggregate rating", y= "Rating count", data= ratings)
```

```python
sns.barplot(x= "Aggregate rating", y= "Rating count",hue= "Rating color", data= ratings)
```

```python
sns.barplot(x= "Aggregate rating", y= "Rating count",hue= "Rating color", data= ratings, palette= ["blue", "red", "orange", "yellow", "green", "green"])
```

Observations:
1. Not Rated count is very high.
2. Maximum number of ratings are between 2.5 and 3.4

```python
sns.countplot(x= "Rating color", data= ratings, palette= ["blue", "red", "orange", "yellow", "green", "green"])
```
