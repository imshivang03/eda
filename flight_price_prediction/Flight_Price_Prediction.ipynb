## Feature Engineering

### Feature Engineering on Date_of_Journey

```python
## Creating 3 derived features for date, month and year of the Date_of_Journey column.

final_df["Date"]= final_df["Date_of_Journey"].str.split("/").str[0]
final_df["Month"]= final_df["Date_of_Journey"].str.split("/").str[1]
final_df["Year"]= final_df["Date_of_Journey"].str.split("/").str[2]

"""
## Applying the lambda function and doing the same split
final_df["Date"]= final_df["Date_of_Journey"].apply(lambda x:x.split("/")[0])
final_df["Month"]= final_df["Date_of_Journey"].apply(lambda x:x.split("/")[1])
final_df["Year"]= final_df["Date_of_Journey"].apply(lambda x:x.split("/")[2])
"""
```

```python
final_df.head(2)
```

```python
## Converting the string into integer

final_df["Date"]= final_df["Date"].astype(int)
final_df["Month"]= final_df["Month"].astype(int)
final_df["Year"]= final_df["Year"].astype(int)
```

```python
final_df.info()
```

```python
## Dropping the Date_of_Journey feature which is not required now because we have splitted that feature into three parts.

final_df.drop("Date_of_Journey", axis=1, inplace=True)
```

### Feature Engineering on Arrival_Time

```python
final_df.head(1)
```

```python
## Splitting the arrival time according to the time and removing date from it.

final_df["Arrival_Time"]= final_df["Arrival_Time"].apply(lambda x:x.split(" ")[0])
```

```python
final_df.head(2)
```

```python
## Splitting the Arrival_Time feature into Arrival_hour and Arrival_min feature.

final_df["Arrival_hour"]= final_df["Arrival_Time"].apply(lambda x:x.split(":")[0])
final_df["Arrival_min"]= final_df["Arrival_Time"].apply(lambda x:x.split(":")[1])
```

```python
final_df.head(2)
```

```python
final_df.info()
```

```python
## Changing the data type of the Arrival_hour and Arrival_min feature from string to integer.

final_df["Arrival_hour"]= final_df["Arrival_hour"].astype(int)
final_df["Arrival_min"]= final_df["Arrival_min"].astype(int)
```

```python
final_df.dtypes
```

```python
## Dropping the Arrival_Time feature which is not required now.

final_df.drop("Arrival_Time", axis=1, inplace= True)
```

```python
final_df.head(1)
```

### Feature Engineering on Dep_Time

```python
## Splitting the Dep_Time feature into Dept_hour and Dept_min feature.

final_df["Dept_hour"]= final_df["Dep_Time"].apply(lambda x:x.split(":")[0])
final_df["Dept_min"]= final_df["Dep_Time"].apply(lambda x:x.split(":")[1])
```

```python
## Changing the data type of the Dept_hour and Dept_min feature from string to integer.

final_df["Dept_hour"]= final_df["Dept_hour"].astype(int)
final_df["Dept_min"]= final_df["Dept_min"].astype(int)
```

```python
## Dropping the Dep_Time feature which is not required now.

final_df.drop("Dep_Time", axis=1, inplace= True)
```

```python
final_df.dtypes
```

```python
final_df.head(1)
```

### Feature Engineering on Total_Stops

```python
final_df["Total_Stops"].unique()
```

```python
final_df[final_df["Total_Stops"].isnull()]
```

```python
## Mapping the values of Total_Stops ("non-stop":0, "1 stop":1, "2 stops":2, "3 stops":3, "4 stops":4, "nan":1)

final_df["Total_Stops"]= final_df["Total_Stops"].map({"non-stop":0, "1 stop":1, "2 stops":2, "3 stops":3, "4 stops":4, "nan":1})
```

```python
final_df.head()
```

```python
## We can now drop the Route because Source, Destination & Total_Stops will be referenced in case of Route.

final_df.drop("Route", axis=1, inplace= True)
```

```python
final_df.head(1)
```

```python
final_df.dtypes
```

### Feature Engineering on Duration

```python
final_df["duration_hour"]= final_df["Duration"].str.split("h").str[0]

# The above code can also be written as:
# final_df["Duration"].str.split(" ").str[0].str.split("h").str[0]
```

```python
final_df.head()
```

```python
## Fetching the values having the duration_hour or Duration = 5m because it can't be converted into integer. 
## This is also a duplicate record because the Duration of the travel from Mumbai to Hyderabad can't be 5m as well as 
## the value is not an integer, so dropping these two rows would be a perfect idea. 

final_df[final_df["duration_hour"]== "5m"]
```

```python
## Dropping the two rows where Duration = 5m

final_df.drop(6474, axis=0, inplace= True)
final_df.drop(2660, axis=0, inplace= True)
```

```python
## Deleted finally.

final_df[final_df["duration_hour"]== "5m"]
```

```python
## Changing the data type of duration_hour to int

final_df["duration_hour"]= final_df["duration_hour"].astype(int)
```

```python
final_df.dtypes
```

```python
final_df.head()
```

```python
## Converting duration_hour into minutes

final_df["duration_hour"]= final_df["duration_hour"]*60
```

```python
## Gathering the minutes information from Duration into a new feature duration_minutes

final_df["duration_minutes"]= final_df["Duration"].str.split("h").str[1].str.split("m").str[0]
```

```python
final_df.head()
```

```python
## Replacing the whitespaces with NAN in duration_minutes. 

final_df["duration_minutes"]= final_df["duration_minutes"].replace(r'^\s*$', np.nan, regex=True)
```

```python
final_df.head()
```

```python
final_df.dtypes
```

```python
final_df[final_df["duration_minutes"].isnull()]
```

```python
## Replacing the Null values with 0.

final_df["duration_minutes"]= final_df["duration_minutes"].fillna(0)
```

```python
final_df.head()
```

```python
## Changing the data type of duration_minutes to int.

final_df["duration_minutes"]= final_df["duration_minutes"].astype(int)
```

```python
final_df.head()
```

```python
## Adding both the duration_hour (after converting into minutes) and duration_minutes and storing into a new column duration.

final_df["duration"]= final_df["duration_hour"] + final_df["duration_minutes"]
```

```python
final_df.head()
```

```python
final_df.dtypes
```

```python
## Dropping the three columns Duration, duration_hour, duration_minutes.

final_df.drop("Duration", axis= 1, inplace= True)
final_df.drop("duration_hour", axis= 1, inplace= True)
final_df.drop("duration_minutes", axis= 1, inplace= True)
```

```python
final_df.head()
```

### Feature Engineering using label encoding on Airline, Source, Destination & Additional_Info

Label Encoding is a popular encoding technique for handling categorical variables. In this technique, each label is assigned a unique integer based on alphabetical ordering.

```python
final_df["Airline"].unique()
```

```python
## Applying label encoding on Airline.
## get_dummies() can also done, since we try to work on train and test data so it's better to use transform techniques.
## Whenever we have test data, we need to transform that test data so that we can save the encoder in the form of 
## pickle file, that's why we shouldn't use get_dummies.

from sklearn.preprocessing import LabelEncoder
labelencoder= LabelEncoder()
```

```python
## Applying label encoding on Airline, Source, Destination, Additional_Info.

final_df["Airline"]= labelencoder.fit_transform(final_df["Airline"])
final_df["Source"]= labelencoder.fit_transform(final_df["Source"])
final_df["Destination"]= labelencoder.fit_transform(final_df["Destination"])
final_df["Additional_Info"]= labelencoder.fit_transform(final_df["Additional_Info"])
```
